## Landscan Pop

```python
import logging

import numpy as np
import geopandas as gpd
import rasterio
from rasterio.mask import mask
from rasterio.enums import Resampling
from shapely.geometry import Polygon
```

```python
# Aggregate the leuven boundary to a single polygon, and then convert it to a GeoDataFrame
leuven_boundary_aggregated = leuven_boundary.unary_union
leuven_boundary_aggregated = gpd.GeoDataFrame(geometry=[leuven_boundary_aggregated])
leuven_boundary_aggregated.crs = leuven_boundary.crs
leuven_boundary_aggregated
```

```python
def filter_population_by_shp(population_tif: str, 
                             boundary_gdf: gpd.GeoDataFrame, 
                             nodata_val=None
                             ):
    """
    Read the population raster and filter the data by the boundary shapefile.
    
    Params:
    ----------
    population_tif: str
        Population raster (.tif) file path
    boundary_gdf: geopandas.GeoDataFrame
        Boundary shapefile in GeoDataFrame format
    nodata_val: float, optional
        The 'nodata' value of the raster, if not specified, it will be read from the metadata
        
    Returns:
    ----------
    band_array: numpy.ndarray
        The filtered population data in numpy array
    out_transform: Affine
        The affine transformation of the filtered raster data 
    crs: Any
        The coordinate reference system of the filtered raster data
    final_nodata: float or None
        The 'nodata' value of the filtered raster data
    """
    
    # Open the raster file
    with rasterio.open(population_tif) as src:
        # Get the 'nodata' value of the raster if not specified
        final_nodata = nodata_val if nodata_val is not None else src.nodata
        
        # Ensure the boundary CRS is the same as the raster CRS
        if boundary_gdf.crs != src.crs:
            boundary_gdf = boundary_gdf.to_crs(src.crs)
        
        # Convert the boundary to GeoJSON format
        logging.info("Converting the boundary to GeoJSON format...")
        shapes = [feature.__geo_interface__ for feature in boundary_gdf.geometry]
        
        # Clip the raster by the boundary
        logging.info("Clipping the raster by the boundary...")
        out_image, out_transform = mask(
            dataset=src, 
            shapes=shapes,
            crop=True,
            nodata=final_nodata,
            all_touched=False   # if True, all pixels touched by the shape will be burned in
        )
        
        band_array = out_image[0]  # Read the first band (population value) of the clipped raster
        logging.warning("Raster has been clipped successfully!")
        return band_array, out_transform, src.crs, final_nodata
```

```python
def grids_to_gdf(band_array, transform, crs, nodata_val=None, min_value=None):
    """
    Convert the clipped population grids (raster) to vector (.shp) format.
    
    Params：
    ----------
    band_array: numpy.ndarray
        2-D array of population values in each grid, which is got from @filter_population_by_shp
    transform: rasterio.transform.Affine
        The affine transformation of the raster
    crs: Any
        The coordinate reference system of the raster
    nodata_val: float or None
        If specified, the nodata value of the raster, which will be excluded from the output
    min_value: float or int, optional
        If specified, the grid with value that is less than the min-value of the population, will be excluded.
        To keep all grids, set it to "-1".
    
    Returns：
    ----------
    gdf: geopandas.GeoDataFrame
        A GeoDataFrame containing each population grid
    """
    
    height, width = band_array.shape
    polygons = []
    records = []
    
    # A for-loop to iterate each grid
    logging.info("Converting the population grids to GeoDataFrame...")
    for row in range(height):
        for col in range(width):
            val = band_array[row, col]
            
            # Exclude the nodata value
            if nodata_val is not None and val == nodata_val:
                continue
            
            # Exclude the value less than the min-value
            if min_value is not None and val <= min_value:
                continue
            
            # Calculate the coordinates of the grid
            x_left, y_top = transform * (col, row)
            x_right, y_bottom = transform * (col + 1, row + 1)
            
            # Construct the polygon
            poly = Polygon([
                (x_left,  y_top),
                (x_right, y_top),
                (x_right, y_bottom),
                (x_left,  y_bottom)
            ])
            
            # Calculate the center of the grid
            x_center = 0.5 * (x_left + x_right)
            y_center = 0.5 * (y_top + y_bottom)
            
            # Construct the record of the grid
            record = {
                "grid_id": f"R{row}C{col}", 
                "row": row,
                "col": col,
                "population": val,
                "x_center": x_center,
                "y_center": y_center
            }
            polygons.append(poly)
            records.append(record)
    
    # Convert the polygons and records to GeoDataFrame
    logging.info("Concat vector grids...")
    gdf = gpd.GeoDataFrame(
        data=records,
        geometry=polygons,
        crs=crs
    )
    print(gdf)
    # If the ''population'' column is not Integer, convert it to Integer
    if gdf["population"].dtype != "int":
        try:
            gdf["population"] = gdf["population"].astype("int")
        except Exception as e:
            logging.error(f"Convert the 'population' column to Integer failed: {e}")

    logging.warning("Population grids have been converted to GeoDataFrame successfully!")
    return gdf
```

```python
def convert_pop_raster2shp(population_tif: str, 
                           boundary_gdf: gpd.GeoDataFrame, 
                           nodata_val=None,
                           min_value=-1,
                           ):
    """
    A wrapper function to convert the population raster to shapefile.
    
    Args:
        population_tif (str): _description_
        boundary_gdf (gpd.GeoDataFrame): _description_
        nodata_val (_type_, optional): _description_. Defaults to None.
        min_value (_type_, optional): _description_. Defaults to -1.

    Returns:
        gdf (gpd.GeoDataFrame): _description_
    """

    # Filter the population raster by the boundary
    band_array, out_transform, crs, final_nodata = filter_population_by_shp(
        population_tif, boundary_gdf, nodata_val
    )

    # Convert the population grids to GeoDataFrame
    gdf = grids_to_gdf(
        band_array, out_transform, crs, final_nodata, min_value
    )

    return gdf
```

# H3 

```python
import h3
```

```python
city_leuven_boundary = gpd.read_file(r'../../../../data/intermediate/test/boundary/CityLeuvenBoundary.geojson')
city_leuven_boundary
```

```python
city_leuven_boundary_aggregated = city_leuven_boundary.unary_union
city_leuven_boundary_aggregated = gpd.GeoDataFrame(geometry=[city_leuven_boundary_aggregated])
city_leuven_boundary_aggregated.crs = city_leuven_boundary.crs
city_leuven_boundary_aggregated
```

```python
# Convert Shapely polygon to coordinates format required by h3
polygon_coords = [list(coord) for coord in city_leuven_boundary_aggregated.to_crs('EPSG:4326').geometry[0].exterior.coords]
# Note: h3 expects coordinates as [lat, lng] pairs
polygon_coords = [[coord[1], coord[0]] for coord in polygon_coords[:-1]]  # Remove last point and swap lat/lng

h3_polygon = h3.LatLngPoly(polygon_coords)
# Get h3 hexagons of the leuven boundary
leuven_boundary_hex = h3.polygon_to_cells(h3_polygon, 9)
leuven_boundary_hex
```

```python
hex_polygons = []
for h in leuven_boundary_hex:
    # Get the coordinates of the hexagon boundary
    boundary = h3.cell_to_boundary(h)
    # Convert to shapely polygon
    polygon = Polygon([(lng, lat) for lat, lng in boundary])
    hex_polygons.append({'h3_index': h, 'geometry': polygon})

# Create GeoDataFrame
hex_gdf = gpd.GeoDataFrame(hex_polygons)
hex_gdf.set_crs(epsg=4326, inplace=True)
hex_gdf.to_crs(epsg=31370, inplace=True)
# Display the first few rows
hex_gdf
```

```python
hex_gdf.to_file(r'../../../../data/intermediate/test/h3/leuven_hex_res9.geojson', driver='GeoJSON')
```
